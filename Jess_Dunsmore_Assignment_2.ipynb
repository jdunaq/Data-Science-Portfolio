###Question 1: CSV Files and Dataframes

```python
#Libraries used
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# a) Importing .csv file
dfmaster = pd.read_csv('AssignmentData.csv')

# b) Size of dfmaster and column names
print("Part B) \nSize of dfmaster is: \n", dfmaster.shape[0], "Rows\n", dfmaster.shape[1], "Columns\n")
print("Columns of dfmaster are:")
#Printing column names
for col in dfmaster.columns:
  print(col)

# c) Creating an expanded dataframe
df_derived = dfmaster
df_derived['Y1^2'] = df_derived.Y1**2
df_derived['Y1^3'] = df_derived.Y1**3
df_derived['|Y2|'] = np.abs(df_derived.Y2)
df_derived['Y1*X'] = df_derived.Y1*df_derived.X
print("\nPart C)\nFirst ten rows of df_derived:\n\n", df_derived.head(10), "\n\n")

# d) Plotting results

A = plt.figure(1)
plt.plot(df_derived.Y1,df_derived.Y2)
plt.title("Y2 vs Y1"), plt.xlabel("Y1"), plt.ylabel("Y2")
A.show()

B = plt.figure(2)
plt.plot(df_derived.Y1,df_derived["Y1^2"])
plt.title("Y1^2 vs Y1"), plt.xlabel("Y1"), plt.ylabel("Y1^2")
B.show()

C = plt.figure(3)
plt.plot(df_derived.Y1,df_derived["Y1^3"])
plt.title("Y1^3 vs Y1"), plt.xlabel("Y1"), plt.ylabel("Y1^3")
C.show()

D = plt.figure(4)
plt.plot(df_derived.Y2,df_derived["|Y2|"])
plt.title("|Y2| vs Y2"), plt.xlabel("Y2"), plt.ylabel("|Y2|")
D.show()

E = plt.figure(5)
plt.plot(df_derived.Y2,df_derived["Y1*X"])
plt.title("Y1*X vs Y2"), plt.xlabel("Y2"), plt.ylabel("Y1*X")
E.show()
```

###Question 2: JSON Files and Dataframes

```python
# JSON file that ranks countries by population size from 2017 (src = https://data.world/edmadrigal/world-population-json)
world_pop = pd.read_json('worldpopulation.json')

# Create a figure with axes
fig, ax = plt.subplots()

fig.suptitle('Ranked World Population by Country (2017)') #Figure title
ax.axis('off') #Remove visibility of figure axes
ax.axis('tight') #Set limits of axes to the extent of the data

# Create a table for the first twenty rows of the JSON file
t = ax.table(cellText=world_pop.head(20).values, colLabels=world_pop.columns, colColours=["palegreen"] * 10, loc='center')

# Enable automatic setting of the fontsize in the table
t.auto_set_font_size(True)

# Display table
plt.show()
```
